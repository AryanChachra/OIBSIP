```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn import metrics
import seaborn as sns
```

```python
df=pd.read_csv('Advertising.csv')
df.head
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['TV'],y=df['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Radio'],y=df['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Newspaper'],y=df['Sales'])
plt.show()
```

```python
train,test=train_test_split(df,test_size=0.30)
```

```python
train_x=train[['TV','Radio','Newspaper']]
train_y=train.Sales

test_x=test[['TV','Radio','Newspaper']]
test_y=test.Sales
```

```python
train_x.head
```

```python
train_y.head
```

```python
model=linear_model.LinearRegression()
model.fit(train_x,train_y)
predct=model.predict(test_x)
metrics.r2_score(predct,test_y)
```

```python
plt.figure(figsize=(10,5))
sns.regplot(x=test_y,y=predct,scatter_kws={'color':'red'})
```
